```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = r"C:\Users\lenovo\Downloads\Q2.xlsx"
df = pd.read_excel(file_path) 

# Display basic info
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Show first few rows
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values with median imputation
df.fillna(df.median(), inplace=True)
```

```python
df["Paid_Search_Lagged"] = df["Paid Search Spends"].shift(3)  
# Marketing campaigns may take 3 weeks to show results
df["Week_Number"] = df["Week_Ending"].dt.isocalendar().week  
# Extract seasonal patterns
df["Month"] = df["Week_Ending"].dt.month
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = ["Discount1", "Discount2", "Gasoline Price", "Average Price", "Paid Search Spends",
            "Paid Social Spends", "Modular Video Spends", "Email Spends", "Organic Search Impressions"]
df[features] = scaler.fit_transform(df[features])
```

```python
import statsmodels.api as sm

# Define independent variables (X) and dependent variable (y)
X = df[["Discount1", "Discount2", "Gasoline Price", "Average Price", "Paid Search Spends",
        "Paid Social Spends", "Modular Video Spends", "Email Spends", "Organic Search Impressions", "Holiday Dummy"]]

y = df["Sales"]

# Add constant for OLS regression
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print model summary (statistical outputs)
print(model.summary())
```

```python
#Interpretation of the Result
#Model Summary:

#R-squared = 0.661  The model explains 66.1% of the variation in Sales, which is fairly strong but leaves room for unexplained variance.
#Adjusted R-squared = 0.631  After adjusting for the number of predictors, the model still maintains a decent explanatory power.
#F-statistic = 21.68, p-value = 7.46e-22  The overall model is statistically significant, meaning at least one predictor has a real impact on Sales.
#Observations = 122  The model is based on 122 data points, which is adequate for regression.
```
